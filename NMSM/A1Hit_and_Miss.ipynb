```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
rectangle = pd.read_csv("https://raw.githubusercontent.com/pevaldes/UNIPD_Physics/main/NMSM/rectangle.txt", sep=";")
print(len(rectangle))
rectangle.head()
```

```python
x_left = float(rectangle.columns[0][2])
x_right = float(rectangle.columns[0][4])
y_top = float(rectangle.columns[1][5])
y_bottom = float(rectangle.columns[1][3])
```

```python
rectangle_plot = sns.scatterplot(data=rectangle, x=rectangle.columns[0], y=rectangle.columns[1])
rectangle_plot.set_xlim(left = x_left-0.25, right = x_right + 0.25)
rectangle_plot.set_ylim(bottom = y_bottom-0.25, top = y_top+0.25)
rectangle_plot.set_title("Area of a rectangle")
```

```python
disk = pd.read_csv("https://raw.githubusercontent.com/pevaldes/UNIPD_Physics/main/NMSM/disk_numbers.txt", sep=";")
disk.head()
```

```python
disk['hit'] = np.where(disk[disk.columns[2]] < float(disk.columns[2][2:-1]), 1, 0)
```

```python
plt.figure(figsize=(7,7))
disk_plot = sns.scatterplot(data=disk, x="X", y="Y", hue='hit')
disk_plot.set_xlim(left = -0.1, right = float(disk.columns[2][2:-1]) + 0.1)
disk_plot.set_ylim(bottom = -0.1, top = float(disk.columns[2][2:-1]) + 0.1)
disk_plot.set_title("Monte Carlo Hit and Miss for a quarter of a unitary disk")
```

```python
disk_left = disk.copy()
disk_left_down = disk.copy()
disk_right_down = disk.copy()
```

```python
disk_left['X'] = -disk['X']
disk_left_down['X'] = -disk['X']
disk_left_down['Y'] = -disk['Y']
disk_right_down['Y'] = -disk['Y']
```

```python
disk_complete = disk.copy()
```

```python
disk_complete = disk_complete.append(disk_left)
disk_complete = disk_complete.append(disk_left_down)
disk_complete = disk_complete.append(disk_right_down)

len(disk_complete)
```

```python
plt.figure(figsize=(7,7))
disk_plot_complete = sns.scatterplot(data=disk_complete, x='X', y='Y', hue = 'hit')
disk_plot_complete.set_xlim(left = -1.1, right = 1.1)
disk_plot_complete.set_ylim(bottom = -1.1, top = 1.1)
disk_plot_complete.set_title("Monte Carlo Hit and Miss for the unitary disk")
```
