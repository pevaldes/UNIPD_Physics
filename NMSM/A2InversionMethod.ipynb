```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
def histogram_plot(n, a):
    """
    n = power
    a = upper limit
    """
    file_name = "distribution_pow{}_lim{}.txt".format(n,a)
    file = open("ExercisesA1A5/{}".format(file_name), 'r')
    data = file.read()
    
    data_list = data.split(";")[:-1]
    data_list = list(map(float,data_list))
    
    c = (n+1.0)/a**(n+1) #Normalization constant
    graph_title = "p(x) = $c.x^n$\n c = {}\n n = {}".format(c,n)
    
    x = np.linspace(0,a,1000)
    f_x = c*x**n
    
    sns.histplot(data_list, stat="density").set(title=graph_title)
    sns.lineplot(x,f_x, color="red", label="p(x) = c.x^n")
    plt.legend(loc=0)
```

```python
fig = plt.figure(figsize=(20,15))

plt.subplot(2,2,1)
histogram_plot(2,2)
```

```python
fig = plt.figure(figsize=(20,15))
plt.subplot(2,2,2)
histogram_plot(3,1)
```
