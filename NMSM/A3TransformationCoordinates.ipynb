```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
direct = pd.read_csv("https://raw.githubusercontent.com/pevaldes/UNIPD_Physics/main/NMSM/direct_sampling_disk.txt", sep=";")
print(len(direct))
direct.head()
```

```python
plt.figure(figsize=(7,7))
direct_plot = sns.jointplot(data=direct,x='x',y='y')
```

```python
transform = pd.read_csv("https://raw.githubusercontent.com/pevaldes/UNIPD_Physics/main/NMSM/transform_sampling.txt", sep=";")
print(len(transform))
transform.head()
```

```python
plt.figure(figsize=(7,7))
transform_plot = sns.jointplot(data=transform,x='x',y='y')
```
